```python
import spacy
from spacy.lookups import Lookups
from spacy.training.example import Example
import random

nlp = spacy.load("en_core_web_sm")

if "textcat" not in nlp.pipe_names:
    textcat = nlp.add_pipe("textcat", last=True)
else:
    textcat = nlp.get_pipe("textcat")

textcat.add_label("POSITIVE")
textcat.add_label("NEGATIVE")

lookups = Lookups()
lookups.add_table("lexeme_norm", {})
nlp.vocab.lookups = lookups


train_data = [
    ("I love this product, it's amazing!", {"cats": {"POSITIVE": 1, "NEGATIVE": 0}}),
    ("This is the worst experience I've ever had.", {"cats": {"POSITIVE": 0, "NEGATIVE": 1}}),
    ("Absolutely fantastic! Highly recommend.", {"cats": {"POSITIVE": 1, "NEGATIVE": 0}}),
    ("I hate this so much.", {"cats": {"POSITIVE": 0, "NEGATIVE": 1}}),
    ("Not bad, could be better.", {"cats": {"POSITIVE": 1, "NEGATIVE": 0}}),
    ("I wouldn't recommend this to anyone.", {"cats": {"POSITIVE": 0, "NEGATIVE": 1}})
]

optimizer = nlp.begin_training()
print("Training the model...")

for epoch in range(20):
    random.shuffle(train_data)
    losses = {}
    for text, annotations in train_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], sgd=optimizer, losses=losses)
    print(f"Losses at iteration {epoch}: {losses}")

nlp.to_disk("pretrained_text_classification_model")
print("\nModel saved to 'pretrained_text_classification_model' directory.")
```
